#### Filtrar elementos en una lista üîç
- Crea una lista con los nombres de tus amigos üßë‚Äçü§ù‚Äçüßë.

## üìö<font color='green'>Diccionarios</font>



### üí° ¬øQu√© son?

Son estructuras de datos que nos permiten almacenar **pares de valores**: una **clave** y un **valor** asociado.

Son muy √∫tiles cuando necesitamos acceder r√°pidamente a un valor a trav√©s de una clave.

Para crear un diccionario en Python, utilizamos llaves y separamos los pares clave-valor por comas.

### Principales operaciones con diccionarios üìïüìù
#### Creaci√≥n üìä
- Crea una lista con los nombres de tus libros favoritos üìöüëÄ.
- Ordena la lista alfab√©ticamente y muestra los libros en orden ascendente üìñüîù.

````python
# Creaci√≥n
mi_diccionario = {'nombre': 'Juan', 'edad': 30, 'ciudad': 'Lima'}
print(mi_diccionario['nombre'])  # salida: 'Juan'
print(mi_diccionario['edad'])  # salida: 30

# Crea un diccionario con los nombres de tus amigos y la ciudad donde viven
amigos = {"Juan": "Bogot√°", "Maria": "Medell√≠n", "Luis": "Cali", "Ana": "Barranquilla"}

# Pide al usuario que agregue un amigo nuevo y la ciudad donde vive
nuevo_amigo = input("Por favor, escribe el nombre de un nuevo amigo: ")
ciudad_nuevo_amigo = input("Por favor, escribe la ciudad donde vive el nuevo amigo: ")
amigos[nuevo_amigo] = ciudad_nuevo_amigo

# Muestra el diccionario completo
print("Este es tu diccionario de amigos y ciudades:")
for amigo, ciudad in amigos.items():
    print(f"- {amigo} vive en {ciudad}")
````

#### Agregar y eliminar elementos en un diccionario üìù
- Crea un diccionario con los nombres de tus pel√≠culas favoritas y su a√±o de lanzamiento üé•üìÖ.
- Pide al usuario que agregue una pel√≠cula nueva con su a√±o de lanzamiento y elimine una pel√≠cula existente.
- Muestra el diccionario completo actualizado ü§ØüëÄ.

````python
# Crea un diccionario con los nombres de tus pel√≠culas favoritas y su a√±o de lanzamiento
peliculas = {"El Padrino": 1972, "Pulp Fiction": 1994, "El Rey Le√≥n": 1994, "Titanic": 1997}

# Pide al usuario que agregue una pel√≠cula nueva con su a√±o de lanzamiento
nueva_pelicula = input("Por favor, escribe el nombre de una nueva pel√≠cula: ")
anio_nueva_pelicula = int(input("Por favor, escribe el a√±o de lanzamiento de la nueva pel√≠cula: "))
peliculas[nueva_pelicula] = anio_nueva_pelicula

# Pide al usuario que elimine una pel√≠cula existente
pelicula_eliminar = input("Por favor, escribe el nombre de una pel√≠cula para eliminarla: ")
del peliculas[pelicula_eliminar]

# Muestra el diccionario completo actualizado
print("Este es tu diccionario de pel√≠culas actualizado y a√±os de lanzamiento:")
for pelicula, anio in peliculas.items():
    print(f"- {pelicula} se lanz√≥ en {anio}")
````

#### Ordenar un diccionario üìä
- Crea un diccionario con los nombres de tus libros favoritos y su autor üìöüßë‚Äçüè´.
- Ordena el diccionario alfab√©ticamente por el nombre del libro y muestra los libros en orden ascendente üìñüîù.

````python
# Crea un diccionario con los nombres de tus libros favoritos y su autor
libros = {"El Quijote": "Miguel de Cervantes", "Cien a√±os de soledad": "Gabriel Garc√≠a M√°rquez", "La Iliada": "Homero", "Orgullo y prejuicio": "Jane Austen"}

# Ordena el diccionario alfab√©ticamente por el nombre del libro
libros_ordenados = dict(sorted(libros.items()))

# Muestra los libros en orden ascendente
print("Estos son tus libros favoritos ordenados alfab√©ticamente:")
for libro, autor in libros_ordenados.items():
    print(f"- {libro}, escrito por {autor}")
````

## üìú<font color='green'>Transformaciones entre listas y diccionarios</font>

Para transformar de listas a diccionarios en Python, puedes usar la funci√≥n **zip()**, que combina dos listas en pares de valores. Luego deber√°s convertir el objeto zip en una **lista** o un **diccionario** utilizando las funciones **list()** o **dict()**, respectivamente. Por ejemplo:

````python
nombres = ["Juan", "Mar√≠a", "Pedro"]
edades = [28, 30, 25]

diccionario_edades = dict(zip(nombres, edades))

print(diccionario_edades)
````

Para transformar de **diccionarios** a **listas**, puedes usar los m√©todos **keys()** y **values()**.
Por ejemplo:

````python
diccionario_edades = {'Juan': 28, 'Mar√≠a': 30, 'Pedro': 25}

nombres = list(diccionario_edades.keys())
edades = list(diccionario_edades.values())

print(nombres)
print(edades)
````

Como vemos en estos ejemplos, para transformar de **diccionario** a **lista** obtendremos o las llaves o los valores del diccionario.

Por otro lado para transformar de **lista** a **diccionario** debemos pasarle pares llave-valor o **tuplas** para adaptar la estructura a la de un **diccionario.**

### üîç Filtrado de un Array

La manera de obtener un **array filtrado** se puede lograr mediante la **indexaci√≥n booleana** de **NumPy**. Por ejemplo, supongamos que tenemos un **array** de **NumPy** llamado arr y queremos obtener un **nuevo array** que contenga solo los valores mayores que 5:

````python
import numpy as np

lista = [1, 6, 3, 8, 2, 9,10]

arr = np.array(lista)

arr[arr > 5]
````

Para filtrar un array por m√°s de una condici√≥n, se puede combinar las condiciones utilizando las funciones:
- **numpy.logical_and().**
- **numpy.logical_or().**

Por ejemplo, si tenemos un arreglo de n√∫meros aleatorios y queremos filtrar los valores que sean mayores a 5 y menores a 10, podemos hacer lo siguiente:

````python
arr[np.logical_and(arr > 5, arr < 10)]
````

## <font color='green'>Carga de datos</font>

Existen varias funciones en NumPy que nos permiten cargar datos desde diferentes fuentes:

- **load():** Esta funci√≥n nos permite cargar un array de NumPy desde un archivo en formato binario (.npy). Este formato es espec√≠fico de NumPy y es muy eficiente en t√©rminos de velocidad de carga y almacenamiento.

- **loadtxt():** Esta funci√≥n nos permite cargar un array de NumPy desde un archivo de texto plano. Los archivos de texto plano son archivos que contienen datos separados por alg√∫n car√°cter (como una coma o un espacio en blanco).

- **genfromtxt():** Esta funci√≥n es similar a loadtxt, pero es m√°s flexible y puede manejar archivos con formatos de datos m√°s complejos

Por ejemplo:

````python
nombres = np.loadtxt("nombre_astronautas.csv", delimiter=',', dtype=str, encoding='utf-8')
print(nombres)
````

Los argumentos que se pasan a la funci√≥n son los siguientes:

- **nombre_astronautas.csv:** este es el nombre del archivo que se desea cargar.

- **delimiter:** este argumento indica que el delimitador utilizado en el archivo es la coma (','). Es decir, cada valor en el archivo est√° separado por comas.

- **dtype:** este argumento indica que se desea cargar los datos como cadenas de texto (strings).

- **encoding:** este argumento indica que el archivo est√° codificado en UTF-8. UTF-8 es un tipo de codificaci√≥n de caracteres que permite representar la mayor√≠a de los caracteres utilizados en el mundo.

Adiiconalmente, podemos utilizar alguns otros par√°metros optativos que nos ser√°n √∫tiles para la carga de archivos separados por coma, como los **csv*:

- **skiprows:** Con este par√°metro puedes indicar el n√∫mero de filas que quieres saltar al leer el archivo. Esto es √∫til si el archivo tiene filas de encabezado que no contienen datos, por ejemplo.

- **usecols:** Este par√°metro te permite seleccionar las columnas que quieres leer del archivo. Puedes pasarle un n√∫mero de columna (empezando por cero) o una lista de n√∫meros si necesitas leer varias columnas.

Por ejemplo, supongamos que tenemos un archivo de datos separados por comas:
a,	b,	c,	d<br>
1,	4,	6,	2<br>
2,	5,	4,	1<br>

4,	8,	0,	3<br>
5,	6,	9,	9<br>
6,	23,	87,	3<br>
7,	8,	6,	75<br>
8,	5,	4,	53<br>
9,	9,	2,	43<br>
10,	6,	2,	3<br>

Utilizamos los par√°metros del m√©todo loadtxt() para indicar que el archivo tiene separaci√≥n por comas, tambi√©n indicamos que solo considere las columnas 1 y 2,  y tambi√©n que se salte la fila que tiene los encabezados.


````python
datos = np.loadtxt('datos.csv', delimiter=',', skiprows=1, usecols=(1, 2), dtype=int)
print(datos)
````

## üöÄ <font color='green'>Estadisticas con NumPy</font>

Podemos utilizar algunas funciones para el c√°lculo de indicadores estad√≠sticos con NumPy:

### üëâüèº Algunos de los m√©todos m√°s √∫tiles de los arrays de NumPy para el c√°lculo de medidas estad√≠sticas son:

- **numpy.mean():** Calcula la media de los elementos del array.
- **numpy.median():** Calcula la mediana de los elementos del array.
- **numpy.std():** Calcula la desviaci√≥n est√°ndar de los elementos del array.
- **numpy.var():** Calcula la varianza de los elementos del array.
- **numpy.percentile():** Calcula el percentil de los elementos del array.

### üßê ¬øPor qu√© son √∫tiles estos m√©todos?

Porque nos permiten entender mejor los datos que estamos analizando y sacar conclusiones m√°s precisas. En estad√≠stica buscamos describir un conjunto de datos, nos apoyaremos en usar estos m√©todos para calcular la media y la desviaci√≥n est√°ndar de algun campo en particular, lo que nos dar√° una idea de c√≥mo est√°n distribuidos los valores.
<hr>

````python
valores = np.array([5,2,7,3,7,4,8,4,2,5,8,4,7,4,7,9,0,5,3,3])
print(np.mean(valores))#Calcula la media de los elementos del array.
print(np.median(valores))#Calcula la mediana de los elementos del array.
print(np.std(valores))#Calcula la desviaci√≥n est√°ndar de los elementos del array.
print(np.var(valores))#Calcula la varianza de los elementos del array.
print(np.percentile(valores,75))#Calcula el percentil 75 de los elementos del array.
````
