```python
## ciclos for y while

## con for / listas imprimir cada color
lista = ['azul','rojo','amarillo','verde','naranjo']

for x in lista:
  print(x)
```

```python
## con while / listas variables imprimir cada color
y=0
while y < len(lista):
  print(lista[y])
  y = y + 1
```
